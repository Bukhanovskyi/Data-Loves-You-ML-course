**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
array1 = np.array([5, 15, 25, 35, 45])
print('Масив зі списку: ', array1)

array2 = np.zeros((1, 7))
print('Масив з нулів розміром 7: ', array2)


array3 = np.ones((1, 6))
print('Масив з одиниць розміром 6: ', array3)

array4 = np.arange(10, 19)
print('Масив з числами від 10 до 19: ', array4)

del array1, array2, array3, array4
```

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2

Виведіть результат виконання кожної операції.

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
data = np.array([18, 27, 33, 41, 52, 68, 79])
print('Оригінальний масив: ', data)

print('Середнє значення: ', np.round(data.mean(), 2) )               
print('Мінімум: ', data.min())                 
print('Максимум: ', data.max())                 
print('Сума: ', data.sum()) 
del data
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
arr = np.array([15, 22, 35, 44, 58, 67])
print('Оригінальний масив: ', arr)

print('Перший елемент: ', arr[0])
print('Останній елемент: ', arr[-1])
print('Елементи з індексами 1 до 3: ', arr[1:3])
print('Елементи більші за 40: ', arr[arr>40])
```

**5. Двовимірні масиви**

Задано матрицю 3x3.

Виконайте наступні дії з функціоналом Numpy (кожен пункт оцінюється в 0.5 бала):
- Виведіть форму (shape) матриці
- Виведіть перший рядок
- Виведіть другу колонку
- Обчисліть суму всіх елементів
- Обчисліть середнє всіх елементів

```python
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])
matrix
```

```python
print('shape матриці', matrix.shape)
print('перший рядок', matrix[0])
print('другa колонку', matrix[:, 1])
print('сумa всіх елементів', matrix.sum())
print('середнє всіх елементів', matrix.mean())
```
